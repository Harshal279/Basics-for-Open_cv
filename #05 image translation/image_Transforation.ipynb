# **Image Transforation**

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy
```

```python
img=cv2.imread('/content/lena.jpg')
```

### **Translation image**

```python
# function for translating img
def translate(img, x, y):
#The second row [0, 1, y] represents the translation along the y-axis (vertical), where y is the amount of translation.
#                                           |
    transmat = numpy.float32([[1, 0, x], [0, 1, y]])
#                                |
#  The first row [1, 0, x] represents the translation along the x-axis (horizontal), where x is the amount of translation
    dimension = (img.shape[1], img.shape[0])# dimension of the input image (width, height)
#                      (width, height)
    return cv2.warpAffine(img, transmat, dimension) # warpAffine is the function for translation
```

```python
translated_img = translate(img, 100, 100)  # Move down 100px and right 100px
#-x --> left
#-y --> up
#+x -->Right
#+y -->down
cv2_imshow(translated_img)
```

### **Rotating the image**

```python
def rotate(img,angle,rotpoint=None):
    (height,width) = img.shape[:2]

    if rotpoint is None :
        rotpoint = (width//2,height//2) # If the rotation point (rotpoint) is not provided, it defaults to the center of the image.
    # getRotationMatrix2D is the function
    # 1.0 is the scaling factor.
    rotMat = cv2.getRotationMatrix2D(rotpoint,angle,1.0)
    dimensions= (width,height)
    return cv2.warpAffine(img,rotMat,dimensions)
```

```python
rotated= rotate(img,45)
# -value -->clockwise
# +value -->anti
cv2_imshow(rotated)
```

### **Resizing image**

```python
Resizing = cv2.resize(img,(1000,1000),interpolation=cv2.INTER_CUBIC)
cv2_imshow(Resizing)
```

### **Flipping the image**

```python
flip = cv2.flip(img,-1)
# 1 --> horizontal flip
# 0 --> horizontal flip
# -1 --> vertical flip
# -0--> vertical flip
cv2_imshow(flip)
```
