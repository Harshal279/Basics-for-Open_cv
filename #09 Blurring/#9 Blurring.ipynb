```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

### Original Image

```python
img= cv.imread('/content/lena.jpg')
cv2_imshow(img)
```

### Average Blur

```python
average = cv.blur(img,(7,7))    # Higher the kernal size greater the blur
cv2_imshow(average)
```

### Gaussian Blur

```python
# Gaussian Blur is more natural then averaging blur
gauss = cv.GaussianBlur(img, (7,7), 0)
cv2_imshow(gauss)       #           |
#                                   |
#                    this is standerd deviation
```

### Median Blur

```python
 # Median Blurring is more effective in blurring then other two for reducing the noise
 median = cv.medianBlur(img,7)
 cv2_imshow(median)
 # This Bluring is not recommended for higher kernal size
```

### Biletral Blur

```python
# Bilertral blur is most effective among all used in large project
Bilertral =  cv.bilateralFilter(img, 5,35,25)
cv2_imshow(Bilertral)
          #         syntax     (image,pixelDiameter,sigmaColor,SpaceSigma )
```
