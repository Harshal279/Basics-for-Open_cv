# Masking
Size of mask should be equal to the image size Masking

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img =cv.imread('/content/lena.jpg')
cv2_imshow(img)
```

```python
blank1 = np.zeros(img.shape[:2],dtype='uint8')
cv2_imshow(blank1)
```

```python
mask = cv.circle(blank1,(img.shape[1]//2,img.shape[0]//2),150,255,-1)
cv2_imshow(mask)
```

```python
masked = cv.bitwise_and(img,img,mask=mask)
cv2_imshow(masked)
```

```python
blank2 = np.zeros(img.shape[:2],dtype='uint8')
```

```python
rectangle = cv.rectangle(blank2.copy() ,(30,30) ,(370,370),225,-1)
circle1 = cv.circle(blank2.copy() ,(img.shape[1]//2,img.shape[0]//2),200,255,-1)
```

```python
Bitwise_and = cv.bitwise_and(rectangle,circle1)
cv2_imshow(Bitwise_and)
```

```python
masked1 = cv.bitwise_and(img,img,mask=Bitwise_and)
cv2_imshow(masked1)
```
