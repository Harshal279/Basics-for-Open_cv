```python
import cv2 as cv
from google.colab.patches import cv2_imshow     # With the help of this library imshow function will work
```

```python
image = cv.imread('lena.jpg')  # Reading the image
```

```python
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)          # Converting the image into gray
cv2_imshow(gray)
```

```python
blur = cv.GaussianBlur(gray,(3,3),cv.BORDER_DEFAULT)    # Here blured image is used to reduce the no of contours
cv2_imshow(blur)
```

```python
# The threshold function works on threshold value as density below it pixel will be 0(black) above it will be bright
ret ,Thresh =cv.threshold(gray,125,225,cv.THRESH_BINARY)       # binary will take only 1 and 0
cv2_imshow(Thresh)
```

```python
canny  = cv.Canny(blur,125, 175)   # converting image into canny
cv2_imshow(canny)
```

```python
# This function will show the no of contours
contours, hierarchies = cv.findContours(Thresh,cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
print (f'{len(contours)}')
```

RETER_TREE -->  ALL HIERARCHIALS COUNTORS

RETER_EXTERNAL  -->  ONLY EXTERNAL COUNTORS

RETER_LIST   --> ALL COUNTORS
