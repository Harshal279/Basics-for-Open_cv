# Types of colour spaces.

## These code demonstrate various color space conversions and image manipulations using OpenCV

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('lena.jpg')
cv2_imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

**WE CANNOT CONVERT GRAY SCALE IMAGE DIRECTILY TO HSV**

```python
#This function converts the color image from the BGR color space to the HSV color space.
HSV = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
cv2_imshow(HSV)
```

```python
#his function displays the inverted image
Inversion = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2_imshow(Inversion)
```
